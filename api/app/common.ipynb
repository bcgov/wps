```python
import base64
from aiohttp import ClientSession

# TODO: get a API key
# TODO: consider self hosting (sla etc.?) - p.o. decision.
# TODO: figure out iterating over data.

with open('./template_carbone_2.docx', 'rb') as f:

    data = {
        "data": {
            "title": "Demonstration of Carbone 2.1.1 Features",
            "mainBlurb": "The underlying document and templating engine employed by the Common Document Generation Service is Carbone. They have updated their library and added some nice features. Here we will demonstrate some of them for you.",
            "showSectionTwo": 1,
            "sectionTwoText": "You should be seeing this if showSectionTwo = 1",
            "this-is-my-field-name": "This is my sentence. There are many like it, but this one is mine.",
            "this_is_another_field_name": "My sentence is my best friend. It is my life. I must master it as I must master my life."
        },
        "formatters": "",
        "options": {
            "reportName": "template_carbone_2",
            "convertTo": "pdf",
            "overwrite": True
        },
        "template": {
            "content": base64.b64encode(f.read()).decode('utf-8'),
            "encodingType": "base64",
            "fileType": "docx"
        }
    }

async with ClientSession() as session:
    # need an API key prod.
    async with session.post('http://localhost:3000/api/v2/template/render', json=data) as response:
        print(response.status)
        with open('./template_carbone_2.pdf', 'wb') as f:
            f.write(await response.read())
        # print(await response.text())
```
