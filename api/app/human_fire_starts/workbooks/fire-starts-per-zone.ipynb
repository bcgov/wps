```python
import geopandas
gdf = geopandas.read_file('../data/PROT_HISTORICAL_INCIDENTS_SP.geojson')
gdf
```

```python
import pandas as pd
# Filter out non-dates, map dates to datetimes, normalize all dates to midnight, convert all dates to unix timestamps
gdf['IGNITION_DATE'] = gdf['IGNITION_DATE'].dropna().map(pd.to_datetime).dt.normalize()
# Group by fire centre, zone and ignition date and get counts
out = gdf[['FIRE_CENTRE', 'ZONE', 'IGNITION_DATE']].groupby(['FIRE_CENTRE', 'ZONE', 'IGNITION_DATE']).size()
# Put groupby dataframe into a new frame without multi index so we get flat rows
out = out.to_frame(name = 'COUNT').reset_index()
out
```
